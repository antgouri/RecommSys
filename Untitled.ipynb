```python
def check(text):
    for lines in text:
        if any(word in lines for word in pos_high):
            rating = 50
    return rating
```

```python
def applyAll(text):
    if any(x in text for x in pos_high):
        rating = 5
    elif any(x in text for x in pos_medium):
        rating = 3
    elif any(x in text for x in pos_low):
        rating = 1 
    else:
        print('No rating')
```
